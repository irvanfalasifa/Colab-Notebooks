**Apakah Fulan akan Diterima atau Ditolak** jika memiliki Kriteria sebagai berikut :
* Agama    : 8
* Rupa     : 7
* Harta    : 3
* Pandai   : 9
* Romantis : 4

```python
jmlh_ds = len(ds)
jmlh_diterima = ds[(ds.Hasil == 'Terima')]['Hasil'].count()
jmlh_ditolak = ds[(ds.Hasil == 'Tolak')]['Hasil'].count()

#Agama yang bernilai 8
Agama_diterima = ds[(ds.Agama == 8) & (ds.Hasil == 'Terima')]['Hasil'].count()/jmlh_diterima
Agama_ditolak = ds[(ds.Agama == 8) & (ds.Hasil == 'Tolak')]['Hasil'].count()/jmlh_ditolak

#Rupa yang bernilai 7
Rupa_diterima = ds[(ds.Rupa == 7) & (ds.Hasil == 'Terima')]['Hasil'].count()/jmlh_diterima
Rupa_ditolak = ds[(ds.Rupa == 7) & (ds.Hasil == 'Tolak')]['Hasil'].count()/jmlh_ditolak

#Harta yang bernilai 3
Harta_diterima = ds[(ds.Harta == 3) & (ds.Hasil == 'Terima')]['Hasil'].count()/jmlh_diterima
Harta_ditolak = ds[(ds.Harta == 3) & (ds.Hasil == 'Tolak')]['Hasil'].count()/jmlh_ditolak

#Pandai yang bernilai 9
Pandai_diterima = ds[(ds.Pandai== 9) & (ds.Hasil == 'Terima')]['Hasil'].count()/jmlh_diterima
Pandai_ditolak = ds[(ds.Pandai == 9) & (ds.Hasil == 'Tolak')]['Hasil'].count()/jmlh_ditolak

#Romantis yang bernilai 4
Romantis_diterima = ds[(ds.Romantis == 4) & (ds.Hasil == 'Terima')]['Hasil'].count()/jmlh_diterima
Romantis_ditolak = ds[(ds.Romantis == 4) & (ds.Hasil == 'Tolak')]['Hasil'].count()/jmlh_ditolak
```

```python
total_diterima = Agama_diterima*Rupa_diterima*Harta_diterima*Pandai_diterima*Romantis_diterima*(jmlh_diterima/jmlh_ds)
total_ditolak = Agama_ditolak*Rupa_ditolak*Harta_ditolak*Pandai_ditolak*Romantis_ditolak*(jmlh_ditolak/jmlh_ds)

print("Fulan yg Diterima : ",total_diterima)
print("Fulan yg Ditolak  : ",total_ditolak)
```

```python
print("  サマリー  ")
print("\r")
if (total_diterima>total_ditolak) : print("Selamat Fulan Diterima")
else : print("Yah Ditolak :( ")
```
