```python
def fungsi_naik(b, a, x):
  if (x <= a):
    nilai = 0
    
  elif (x > a and x < b):
    nilai = (x-a) / (b-a)
  elif (x >= b):
    nilai = 1
  return nilai
```

```python
def fungsi_turun(b, a, x):
  if (x <= a):
    nilai = 1
  elif(x > a and x < b):
    
    nilai = (b-x)/(b-a)
  elif (x >= b):
    nilai = 0
  return nilai
```

```python
def fungsi_agregasi_naik(b, a, alfa):
  return alfa*(b-a)+a
```

```python
def fungsi_agregasi_turun(b, a, alfa):
  return b-(alfa*(b-a))
```

```python
komponen = ["permintaan", "persediaan", "produksi"]
aturan = {
    0: ["permintaan_turun", "persediaan_naik", "produksi_turun"],
    1: ["permintaan_turun", "persediaan_turun", "produksi_turun"],
    2: ["permintaan_naik", "persediaan_naik", "produksi_naik"],
    3: ["permintaan_naik", "persediaan_turun", "produksi_naik"],
}
```

```python
dataset = {}

for data in komponen:
  print("Masukkan nilai ", data)
  nilai_naik = int(input("Nilai {0} naik = ".format(data)))
  nilai_turun = int(input("Nilai {0} turun = ".format(data)))
  dataset.update({data + "_naik": nilai_naik, data + "_turun": nilai_turun})

print(dataset)
```

```python
pertanyaan = {}

for i in range(2):
  nilai = int(input("Nilai awal Variabel {0}  = ".format(komponen[i])))
  pertanyaan.update({komponen[i]: nilai})

print(pertanyaan)
```

```python
nilai_keanggotaan = {}

for data in pertanyaan:
  naik = fungsi_naik(dataset[data+"_naik"], dataset[data+"_turun"], pertanyaan[data])
  turun = fungsi_turun(dataset[data+"_naik"], dataset[data+"_turun"], pertanyaan[data])
  nilai_keanggotaan.update({data+"_naik": naik, data+"_turun": turun})

print(nilai_keanggotaan)
```

```python
alfa = []
z = []

for key, value in aturan.items():
  kondisi1 = nilai_keanggotaan[value[0]]
  kondisi2 = nilai_keanggotaan[value[1]]
  
  a = min(kondisi1, kondisi2)
  alfa.append(a)

  if (value[2] == "produksi_turun"):
    zz = fungsi_agregasi_turun(dataset["produksi_naik"], dataset["produksi_turun"], a)
  elif (value[2] == "produksi_naik"):
    zz = fungsi_agregasi_naik(dataset["produksi_naik"], dataset["produksi_turun"], a)
  z.append(zz)

print(alfa)
print(z)
```

```python
df = 0

for i in range(len(alfa)):
  df += alfa[i]*z[i]

defuz = int(df/sum(alfa))
print(defuz)
```
